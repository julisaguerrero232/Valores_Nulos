```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Cargamos el Data Frame
df= pd.read_csv("Vienna.csv")
df
```

```python
# Lista de host names que deseas filtrar
host_names = ["Hannes", "Elxe", "Peter", "Ingela" , "Lucia"]

# Aplicar el filtro
filtro1 = df[df["host_name"].isin(host_names)]
filtro1
filtro1.to_csv("host_name.csv")
```

```python
#Los regostros de los Host que se hallan unido a Airbnd antes del año 2019
filtro2=df[df["host_since"] < "2011-03-03"]
filtro2
filtro2.to_csv("host_since.csv")
```

```python
#Host que responden a más tardar en 1 dias
filtro3= df[(df["host_response_time"] == "within a day")] 
filtro3
filtro3.to_csv("within a day_time.csv")
```

```python
#Registros que tengan al menos 5 accommodates
filtro4=df[df["accommodates"] >= 5]
filtro4
filtro4.to_csv("accommodates.csv")
```

```python
#Registros con tipo de cuarto en Entire home y Private room, FILTRAMOS LA VARIABLE "room_type"
filtro5 = df[df["room_type"].isin(["Private room", "Entire home/apt"])]
filtro5
filtro5.to_csv("room_type.csv")
```

```python
#filtros que a lo mucho cuenten con 4 camas
filtro6=df[df["beds"] <= 4]
filtro6
filtro6.to_csv("camas.csv")
```

```python
#Los registros que tengan 1 baño o que sea "superhost" pero con la condicion de que sean "instant_bookable"
filtro7=df[((df["bathrooms"] == 1) | (df["host_is_superhost"] == "t")) & (df["instant_bookable"] == "t")]
filtro7
filtro7.to_csv("filtro7.csv")
```

```python
#Los registros cuyo review_scores_cleanliness sea menor a 3 Y review_scores_rating sea mayor a 3
filtro8=df[(df["review_scores_cleanliness"] < 3) & (df["review_scores_rating"] > 3)]
filtro8
filtro8.to_csv("registros_condicion.csv")
```
